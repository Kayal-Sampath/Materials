```python
def maximum (a, b):
   big = a
   if b > big:
      big = b
   return big

u, v = 5, 2 # tuple assignment
m = maximum (u, v)
print (m)
```

```python
def order (a, b):
    # a ? b
    if a > b:
        a, b = b, a # exchange a and b
    # a <= b
u, v = 5, 2
print (u, v) # 5 2
order (u, v)
print (u, v) # 2 5
```

```python
def order (a, i, j):
   if a[i] > a[j]:
      a[i], a[j] = a[j], a[i]
   # a[i]  <= a[j]

u = [5, 2]
print (u) # [5, 2]
order (u, 0, 1)
print (u) # [2, 5]
```

```python
a = [5, 8, 2]
order (a, 0, 1)
order (a, 1, 2)
# order (a, 0, 1)
print (a)
```

```python
#%%
stocks = []
f = open("portfolio.dat")
for line in f.readlines():
    fields = line.split()
    record = {
        'name'   : fields[0],
        'shares' : int(fields[1]),
        'price'  : float(fields[2])
    }
    stocks.append(record)
f.close()
```

```python
#%%
prices = {}
for line in open("prices.dat"):
    fields = line.split(',')
    prices[fields[0]] = float(fields[1])
initial = sum([s['shares']*s['price']
                 for s in stocks])
current = sum([s['shares']*prices[s['name']]
                    for s in stocks])
print ("Initial value", initial)
print ("Current value", current)
print ("Gain %10.2f" % (current - initial))
```
